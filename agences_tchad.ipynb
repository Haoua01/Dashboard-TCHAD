```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
from geopy.distance import great_circle

#define the coordinates of each region having agencies
coordinates_regions_with_agencies = {
    "N'Djamena": (12.1347, 15.0554),
    "Ouaddai": (13.8292, 20.8324),
    "Logone Occidental": (8.5667, 16.0833),
    "Moyen-Chari": (9.1429, 18.3923),
    "Salamat": (10.2833, 15.3667),
    "Mayo-Kebbi Ouest": (7.3500, 16.4667),
    "Guéra": (11.0000, 19.5000),
    "Logone Oriental": (8.6500, 16.8500),
    "Mayo-Kebbi Est": (8.9667, 17.5333),
    "Wadi Fira": (17.2833, 21.5833),
    "Lac": (13.1667, 14.1167),
    "Kanem": (13.9667, 13.6667),
}

#define the coordinates of the other regions
coordinates_regions_without_agencies = {
    "Barh el Gazel": (13.5000, 20.0000),
    "Batha": (13.2833, 19.6833),
    "Borkou": (19.7500, 18.6667),
    "Chari-Baguirmi": (11.2400, 16.1000),
    "Ennedi Est": (16.4667, 22.4833),
    "Ennedi Ouest": (17.5833, 21.5000),
    "Hadjer-Lamis": (12.8500, 15.4167),
    "Mandoul": (8.5667, 17.2500),
    "Sila": (11.5167, 21.3333),
    "Tandjilé": (9.1333, 16.1000),
    "Tibesti": (21.8333, 17.0000),
}

#combine the two dictionaries
coordinates = {**coordinates_regions_with_agencies, **coordinates_regions_without_agencies}

# Compute distances
neighbors = {}

for city1, coords1 in coordinates.items():
    neighbors[city1] = {}
    for city2, coords2 in coordinates.items():
        if city1 != city2:
            distance = great_circle(coords1, coords2).kilometers
            neighbors[city1][city2] = distance
            #print(f"{city1} : {city2} = {distance:.2f}")

"""ISIBF score for each city"""

# Function to calculate ISIBF

def calculate_isibf1(num_agencies, neighbors, alpha):
    isibf_values = {}
    for city in num_agencies:
        total = 0
        for neighbor, distance in neighbors[city].items():
            if distance > 0:
                total += np.log2(num_agencies[neighbor] + 1) / alpha**distance
        total+= np.log2(num_agencies[city] + 1)
        isibf_values[city] = total
    return isibf_values

alpha=1.1
isibf_values = calculate_isibf1(region_counts, neighbors, alpha)
isibf_values
```
