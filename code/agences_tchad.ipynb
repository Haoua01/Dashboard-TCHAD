```python
#create a dictionary to store the number of agencies in each region and attribute 0 for regions without agencies
region_counts = Counter()
for city, count in city_counts.items():
    region = city_to_region.get(city, "Unknown")
    region_counts[region] += count

for region in city_to_region.values():
    if region not in region_counts:
        region_counts[region] = 0

print(region_counts)
    


```

```python


"""ISIBF score for each city"""

# Function to calculate ISIBF

def calculate_isibf1(num_agencies, neighbors, alpha):
    isibf_values = {}
    for city in num_agencies:
        total = 0
        for neighbor, distance in neighbors[city].items():
            if distance > 0:
                total += np.log2(num_agencies[neighbor] + 1) / alpha**distance
        total+= np.log2(num_agencies[city] + 1)
        isibf_values[city] = total
    return isibf_values

alpha=1.01
isibf_scores0 = calculate_isibf1(region_counts, neighbors, alpha)

# min-max normalization

def min_max_normalize(values):
    min_value = min(values.values())
    max_value = max(values.values())
    normalized_values = {}
    for key, value in values.items():
        normalized_values[key] = (value - min_value) / (max_value - min_value)
        print(f"{key}: {normalized_values[key]:.2f}")
    return normalized_values

isibf_scores = min_max_normalize(isibf_scores0)

#apply .2f to the values of the dictionary
isibf_scores = {k: round(v, 2) for k, v in isibf_scores.items()}

```
